```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
dataset= load_iris()
```

```python
print(dataset.DESCR)
```

```python
# Dependent and Independent features

X= dataset.data
```

```python
y=dataset.target
```

```python
y
```

```python
## Do not perform standard scaling for decision tree algorithm 
```

```python
## train test split

from sklearn.model_selection import train_test_split

X_train,X_test,y_train, y_test= train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
X_train
```

## Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier= DecisionTreeClassifier(criterion="entropy")
# for small datsets we can use the criterion as "entropy"
```

```python
classifier.fit(X_train,y_train)
```

```python
# vissualizing the decision tree 

from sklearn import tree

plt.figure(figsize=(12,10))
tree.plot_tree(classifier, filled=True)
```

## Post pruning

```python
# As we can see in the level-2 is considered to prune the tree. Left side green block represents max values (36) are
# considered and right side block has max values as 38. The model automatically predicts the highest mentioned record so we 
# will prune the tree at level-2 (max_depth=2) to avoid overfitting problem.
```

```python
# post pruning --> we are pruning the tree after constructing tree
```

```python
# post-pruning

classifier_post= DecisionTreeClassifier(criterion="entropy",max_depth=2)
# for small datsets we can use the criterion as "entropy"
```

```python
classifier_post.fit(X_train,y_train)
```

```python
tree.plot_tree(classifier_post,filled=True)
```

```python
y_pred= classifier_post.predict(X_test)
```

```python
y_pred
```

```python
# Let's use the metrics of classification problem

from sklearn.metrics import accuracy_score, classification_report,confusion_matrix

print("classification report:", classification_report(y_pred,y_test))
print("confusion matrix:", confusion_matrix(y_pred,y_test))
print("accuracy score:", accuracy_score(y_pred,y_test))
```

## Decision tree Pre-pruning and hyper parameter tuning for huge data

```python
## we will get more warning when we perform the pre-pruning operation so filter the warnings

import warnings

warnings.filterwarnings("ignore")
```

```python
# Create a dictionary of parameters

parameters={
    "criterion":['gini', 'entropy', 'log_loss'],
    "splitter":['best', 'random'],
    "max_depth":[1,2,3,4,5],
    "max_features":["auto","sqrt", "log2"]
}
```

```python
from sklearn.model_selection import GridSearchCV

classifier_pre= DecisionTreeClassifier()


clf=GridSearchCV(estimator=classifier_pre,param_grid=parameters,cv=5,scoring="accuracy",refit=True)

#for the binary and multi classification we use scoring="accuracy"
```

```python
clf
```

```python
clf.fit(X_train,y_train)
```

```python
clf.param_grid
```

```python
clf.return_train_score
```

```python
clf.best_params_
```

```python
classifier_pre= DecisionTreeClassifier(criterion='log_loss',max_depth=5,max_features='sqrt',splitter='random')
```

```python
classifier_pre.fit(X_train,y_train)
```

```python
y_pred_pre= classifier_pre.predict(X_test)
```

```python
# Let's use the metrics of classification problem

from sklearn.metrics import accuracy_score, classification_report,confusion_matrix

print("classification report:", classification_report(y_pred_pre,y_test))
print("confusion matrix:", confusion_matrix(y_pred_pre,y_test))
print("accuracy score:", accuracy_score(y_pred_pre,y_test))
```
